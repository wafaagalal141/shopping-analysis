```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
t=sns.load_dataset("titanic")
t.head()
```

```python

sns.countplot(x='survived', data=t,color="pink")
plt.title("Survival Count")
plt.show()
```

```python
sns.boxplot(x='survived', y='age', data=t)
plt.title("Age vs Survival")
plt.show()
```

```python
import seaborn as sns

sns.countplot(x='alone', hue='survived', data=t, palette=['pink','blue'])
plt.title("Survival vs Alone")
plt.xlabel("Alone")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='deck', hue='survived', data=t, palette=['red','green'])
plt.title("Survival by Deck")
plt.xlabel("Deck")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    x='age',
    y='fare',
    hue='survived',
    data=t,
    palette={0:'pink', 1:'#800020'},
    alpha=0.7
)
plt.title("Scatter Plot of Age vs Fare by Survival")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
sns.violinplot(x='survived', y='fare', data=t, palette=['purple','pink'])
plt.title("Fare Distribution by Survival")
plt.show()
```

```python
class_survived = t.groupby('pclass')['survived'].value_counts(normalize=True).unstack()
class_survived.plot(kind='bar', stacked=True, color=['pink','#800020'], figsize=(10,6))
plt.title("Stacked Bar of Survival by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Proportion")
plt.show()
```
